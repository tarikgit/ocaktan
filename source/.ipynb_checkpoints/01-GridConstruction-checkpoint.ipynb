# Grid construction

We explore he three different approaches for constructing a grid in a state space. 
1. Equidistant grid
1. Random grid
1. Chebyshev nodes




```julia
# Construction of the grid: equidistant
dXss = 5.55;
iGridPoints = 3;
dXmin = dXss * 0.8;
dXmax = dXss * 1.2;
iStepX = (dXmax-dXmin) / (iGridPoints-1);
vX = collect(dXmin:iStepX:dXmax);
iGridPoints = size(vX,1);
print(vX)
```

```julia
# Construction of the grid: random
using Random, Distributions
iN = 1; # Number of draws of 'iGridPoints' points
dXss = 5.55;
iGridPoints = 3;
dXmin = dXss * 0.8;
dXmax = dXss * 1.2;
vXtemp = rand(Uniform(dXmin,dXmax),iGridPoints,iN)

vX = sort!(vXtemp, dims=1) 

display( vX )

```

```julia
# Check how many iterations it will take to cover a large part of the state space (limited to 2 decimals)
for i=1:iGridPoints
    for j=1:iN
        vX[i,j] = round(vX[i,j], digits=2);
    end
end
vDiff = vX[iGridPoints,:]-vX[1,:];
display(vDiff)
```

```julia
# Plot the random grid points
using Plots
scatter(vX, ones(iGridPoints,1), title = "Gridpoints")
```
